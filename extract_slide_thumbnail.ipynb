```python
from pdf2image import convert_from_path
import os

# Define paths
pdf_path = "assets/full_deck/end_to_end_pipeline.pdf"
output_folder = "assets/slide_preview"

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Convert PDF pages to images
pages = convert_from_path(pdf_path, dpi=150)

# Save each page as a JPG inside assets/slide_preview
for i, page in enumerate(pages, 1):
    output_path = os.path.join(output_folder, f"recommender_recwalk_page_{i}.jpg")
    page.save(output_path, "JPEG")

print(f"✅ PDF successfully converted to images! Saved to: {output_folder}")
```
